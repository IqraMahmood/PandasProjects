```python
import pandas as pd

categories = pd.read_csv("Product Categories.csv")
```

```python
sku= pd.read_csv("SKU.csv")
```

```python
df = pd.merge(sku, categories, on="product_id")
```

```python
merge = df[['sku_id','product_id','Category ID']]
merge.to_csv('merge.csv', index=False)
```
